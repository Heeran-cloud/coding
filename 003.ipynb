```python
station_names = ['혜화', '남대문', '중부', '종암', '서부', '방배', '수서']
gu_names = ['종로구', '중구', '중구', '성북구', '은평구', '서초구', '강남구']

tmp_gu = []
for idx, _ in crime_raw_data.iterrows():
    if idx in station_names:
        tmp_gu.append(gu_names[station_names.index(idx)])
        
    else:
        tmp_gu.append(idx+'구')
```

```python
tmp_gu
```

```python
crime_raw_data['구'] = tmp_gu
crime_raw_data
```

```python
crime_gu = pd.pivot_table(crime_raw_data, index=['구',], aggfunc=np.sum)
crime_gu
```

```python
num = ['강간검거', '강도검거', '살인검거', '절도검거', '폭력검거']
den = ['강간발생', '강도발생', '살인발생', '절도발생', '폭력발생']

crime_gu[num].div(crime_gu[den].values)
```

```python
target = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
num = ['강간검거', '강도검거', '살인검거', '절도검거', '폭력검거']
den = ['강간발생', '강도발생', '살인발생', '절도발생', '폭력발생']

crime_gu[target] = crime_gu[num].div(crime_gu[den].values) * 100
crime_gu
```

```python
crime_gu.drop(columns=num, inplace=True)
crime_gu
```

```python
crime_gu[crime_gu[target] > 100] = 100
crime_gu
```

```python
crime_gu.rename(columns = {'강간발생':'강간',
                          '강도발생':'강도',
                          '살인발생':'살인',
                          '절도발생':'절도',
                          '폭력발생':'폭력'},
               inplace=True)
crime_gu
```

```python
# 모든 종류의 범죄를 정규화하여 최대값을 1로 만들기
col = ['강간', '강도', '살인', '절도', '폭력']
crime_gu_norm = crime_gu[col] / crime_gu[col].max()
crime_gu_norm
```

```python
col2 = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
crime_gu_norm[col2] = crime_gu[col2]
crime_gu_norm
```

```python
col = ['강간', '강도', '살인', '절도', '폭력']
crime_gu_norm['범죄'] = np.mean(crime_gu_norm[col], axis=1)
```

```python
crime_gu_norm
```

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x='day', y='total_bill', data=tips, color='.5')
plt.show()
# 박스플롯에 비해서 x축 개별적으로 어느 요일에 수요가 몰려있었는지를 알 수 있었고(히스토그램의 역할),
# 빌 어마운트중에서도 어느 금액에 몰려있는지도 알 수 있었다. e.g.)토요일은 10불 후반-20불초반대에 주문량이 가장 많았다. (최빈값)
```
