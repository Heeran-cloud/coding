```python
# when the image is phone-pictured one.
reader = easyocr.Reader(['ko'], gpu=False) # CPU 사용
result = reader.readtext('./참조/phone_pictured.jpg', detail=0)
[each for each in result]
```

```python
output = [each for each in result]
output
```

```python
output[1:8] = ['횟 수', '면  적', '금액(원)', '비고', '번호', '시 설 명']
```

```python
rows = output[7:]
rows
```

```python
import pandas as pd
df = pd.DataFrame(columns=output[1:7])
df
```

```python
row = []
idx = 0
for each in output[7:]:
    row.append(each)
    if len(row) == 5:
        row.append(' ')
        df.loc[idx] = row
        row = []
        idx +=1
df
```

```python
df.to_excel('easyocr_phone_pic_slack.xlsx', encoding='utf-8')
```
